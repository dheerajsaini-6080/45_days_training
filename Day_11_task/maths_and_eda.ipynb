```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return x**2 + 2*x + 1

def derivative(x, h=1e-5):
    return (f(x+h)-f(x))/h

points = [0, 1, 3]

for p in points:
    print(
        f"Derivative at x={p}:",
        derivative(p)
    )
```

```python
x = np.linspace(-5,5,200)

plt.figure(figsize=(8,5))
plt.plot(x,f(x),label="f(x)")

for p in points:

    slope = derivative(p)

    tangent = (
        f(p)
        + slope*(x-p)
    )

    plt.plot(
        x,
        tangent,
        linestyle="--",
        label=f"Tangent at {p}"
    )

plt.legend()
plt.grid()
plt.show()
```

Observation 1:
Slope increases as x increases.

Observation 2:
Positive slope means the function is increasing.

Observation 3:
Near-zero slope indicates we are close to an optimum point.
