```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python

uniform = np.random.uniform(
    0,
    1,
    1000
)

normal = np.random.normal(
    0,
    1,
    1000
)
```

```python
print(
    "Uniform Mean:",
    uniform.mean()
)

print(
    "Uniform Std:",
    uniform.std()
)

print(
    "Normal Mean:",
    normal.mean()
)

print(
    "Normal Std:",
    normal.std()
)
```

```python
plt.hist(uniform,bins=30)

plt.title(
    "Uniform Distribution"
)

plt.show()
```

```python
plt.hist(normal,bins=30)

plt.title(
    "Normal Distribution"
)

plt.show()
```

Uniform distribution spreads values evenly.

Normal distribution concentrates values around the mean.
