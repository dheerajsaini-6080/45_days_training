```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def loss(theta):
    return (theta-3.5)**2 + 1

def grad(theta):
    return 2*(theta-3.5)
```

```python
theta = 10

lr = 0.1

history = []

for i in range(15):

    g = grad(theta)

    l = loss(theta)

    history.append(
        [i,theta,l,g]
    )

    theta = theta - lr*g

df = pd.DataFrame(
    history,
    columns=[
        "Iteration",
        "Theta",
        "Loss",
        "Gradient"
    ]
)

df
```

```python
plt.plot(df["Iteration"],df["Loss"])
plt.title("Loss Curve")
plt.xlabel("Iteration")
plt.ylabel("Loss")
plt.grid()
plt.show()
```

Lower learning rate converges slowly.

Higher learning rate converges faster but may overshoot.
